```python
%matplotlib notebook

import matplotlib.pyplot as plt
import numpy as np

```

```python
import pandas as pd
import numpy as np

np.random.seed(12345)

df = pd.DataFrame([np.random.normal(32000,200000,3650), 
                   np.random.normal(43000,100000,3650), 
                   np.random.normal(43500,140000,3650), 
                   np.random.normal(48000,70000,3650)], 
                  index=[1992,1993,1994,1995])
df
```

```python
plt.figure()
plot= plt.bar(df.index, df.mean(axis=1), yerr= (df.sem(axis=1)*1.96), capsize=10)
plt.xticks(df.index)
plt.title("Interactive bar plot")
hline = plt.axhline()

def evento(event):
    plt.title("Interactive bar plot with y = {} hline coordinates".format(event.ydata))
    for i in range(0, len(df.index)):
        if event.ydata - df.mean(axis=1).tolist()[i] > 2000:
            plot[i].set_color("blue")
        elif event.ydata - df.mean(axis=1).tolist()[i] < -2000:
            plot[i].set_color("red")
        elif event.ydata - df.mean(axis=1).tolist()[i] >= -2000 and event.ydata - df.mean(axis=1).tolist()[i] <= 2000:
            plot[i].set_color("green")
    hline.set_ydata(event.ydata)
    
plt.gcf().canvas.mpl_connect('button_press_event', evento)
```
